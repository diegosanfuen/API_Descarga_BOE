```python
import tkinter as tk
from tkinterhtml import HtmlFrame

def on_mousewheel(event):
    canvas.yview_scroll(-1 * int((event.delta / 120)), "units")

# Crear la ventana principal
root = tk.Tk()
root.title("Etiqueta HTML con Scrollbar")

# Crear un contenedor para la etiqueta HTML
container = tk.Frame(root)
container.pack(fill="both", expand=True)

# Crear una etiqueta HTML dentro del contenedor
html_label = HtmlFrame(container)
html_label.pack(fill="both", expand=True)

# Contenido HTML para la etiqueta
html_content = """
<!DOCTYPE html>
<html>
<head>
<title>HTML Label</title>
</head>
<body>
<h1>Contenido HTML</h1>
<p>Este es un ejemplo de contenido HTML dentro de una etiqueta HTMLLabel en Tkinter.</p>
<p>Esto puede incluir texto formateado, imágenes y otros elementos HTML.</p>
<p>Esta etiqueta se puede desplazar utilizando la barra de desplazamiento.</p>
</body>
</html>
"""

# Establecer el contenido HTML de la etiqueta
html_label.set_content(html_content)

# Crear una barra de desplazamiento
scrollbar = tk.Scrollbar(container, orient="vertical", command=html_label.yview)
scrollbar.pack(side="right", fill="y")

# Asociar la barra de desplazamiento con la etiqueta HTML
html_label.configure(yscrollcommand=scrollbar.set)

# Habilitar el desplazamiento con la rueda del ratón
html_label.bind("<MouseWheel>", on_mousewheel)

# Ejecutar el bucle principal
root.mainloop()
```
